```python
import pandas as pd
from itertools import product
```

```python
returns_df = pd.read_excel("/content/Returns.xlsx", sheet_name=0, index_col=0, engine="openpyxl")
marketcap_df = pd.read_excel("/content/Marketcap.xlsx", sheet_name=0, index_col=0, engine="openpyxl")
```

```python
returns_df.columns = [col.split(" ")[0] for col in returns_df.columns if col != "Date"]
marketcap_df.columns = [col.split(" ")[0] for col in marketcap_df.columns if col != "Date"]
```

```python
# Ensure both datasets have the same tickers
common_tickers = list(set(returns_df.columns) & set(marketcap_df.columns))
```

```python
returns_df = returns_df[common_tickers]
marketcap_df = marketcap_df[common_tickers]
```

```python
# Convert market cap data to numeric values
marketcap_series = marketcap_df.iloc[-1].dropna()
```

```python
marketcap_series
```

```python
# Compute the 20th and 80th percentile values for market capitalization
marketcap_20th_percentile = marketcap_series.quantile(0.25)
marketcap_80th_percentile = marketcap_series.quantile(0.75)

# Print results
print(f"20th Percentile Market Cap: {marketcap_20th_percentile}")
print(f"80th Percentile Market Cap: {marketcap_80th_percentile}")
```

```python
# Define large-cap and small-cap stocks (threshold: $10 billion)
large_cap = marketcap_series[marketcap_series >= 80e3].index.tolist()
small_cap = marketcap_series[marketcap_series < 20e3].index.tolist()
```

```python
# Reduce dataset size by selecting a shorter date range (last 500 days)
returns_df = returns_df.iloc[-500:]
```

```python
# Compute correlation matrix for all stock pairs
correlation_matrix = returns_df.corr()
```

```python
correlation_matrix
```

```python
# Define correlation thresholds
correlation_high_threshold = 0.7
correlation_low_threshold = 0.3
```

```python
def find_mixed_correlation_pairs(large_cap, small_cap, corr_matrix, high=True):
    """Finds pairs of one large-cap and one small-cap stock that are either highly correlated or uncorrelated."""
    pairs = []
    for large, small in product(large_cap, small_cap):
        if large in corr_matrix.index and small in corr_matrix.columns:
            corr = corr_matrix.loc[large, small]
            if high and corr >= correlation_high_threshold:
                pairs.append((large, marketcap_series[large], small, marketcap_series[small], corr))
            elif not high and corr <= correlation_low_threshold:
                pairs.append((large, marketcap_series[large], small, marketcap_series[small], corr))
    return pairs
```

```python
# Find highly correlated and uncorrelated mixed-cap pairs
high_corr_pairs = find_mixed_correlation_pairs(large_cap, small_cap, correlation_matrix, high=True)
low_corr_pairs = find_mixed_correlation_pairs(large_cap, small_cap, correlation_matrix, high=False)
```

```python
# Convert to DataFrames
high_corr_df = pd.DataFrame(high_corr_pairs, columns=["Large Cap Stock", "Large Cap Market Cap", "Small Cap Stock", "Small Cap Market Cap", "Correlation"])
low_corr_df = pd.DataFrame(low_corr_pairs, columns=["Large Cap Stock", "Large Cap Market Cap", "Small Cap Stock", "Small Cap Market Cap", "Correlation"])
```

```python
# Save results to CSV
high_corr_df.to_csv("highly_correlated_mixed_cap_pairs_set.csv", index=False)
low_corr_df.to_csv("uncorrelated_mixed_cap_pairs_set.csv", index=False)

# Print confirmation
print("Mixed large-cap and small-cap pair correlation analysis complete. Results saved as CSV files.")
```
