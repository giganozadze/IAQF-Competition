```python
def create_sequences(X_df, Y_df, seq_length=30):

    X_arr = X_df.values
    Y_arr = Y_df.values

    X_list, Y_list = [], []
    for i in range(len(X_df) - seq_length):
        x_seq = X_arr[i : i+seq_length]    # shape: (seq_length, num_features)
        y_seq = Y_arr[i + seq_length]      # shape: (10,)
        X_list.append(x_seq)
        Y_list.append(y_seq)

    return np.array(X_list), np.array(Y_list)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model, Input
```

```python
class TransformerBlock(layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim, dropout=0.1):
        super().__init__()
        self.att = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = tf.keras.Sequential([
            layers.Dense(ff_dim, activation='relu'),
            layers.Dense(embed_dim),
        ])
        self.norm1 = layers.LayerNormalization(epsilon=1e-6)
        self.norm2 = layers.LayerNormalization(epsilon=1e-6)
        self.dropout1 = layers.Dropout(dropout)
        self.dropout2 = layers.Dropout(dropout)

    def call(self, x, training=None):
        # Multi-head self-attention
        attn_output = self.att(x, x, x)
        attn_output = self.dropout1(attn_output, training=training)
        out1 = self.norm1(x + attn_output)

        # Feed-forward
        ffn_output = self.ffn(out1)
        ffn_output = self.dropout2(ffn_output, training=training)
        return self.norm2(out1 + ffn_output)
```

```python
def build_deeper_transformer(seq_len=30,
                             input_dim=80,
                             hidden_dim=128,
                             num_heads=4,
                             ff_dim=256,
                             num_blocks=3,
                             output_dim=10,
                             dropout=0.1):
    """
    seq_len   : 30  (time steps)
    input_dim : 80  (features per time step)
    hidden_dim: dimension inside the transformer (128 recommended to store more capacity)
    ff_dim    : dimension in feed-forward layers
    num_blocks: number of transformer blocks stacked
    output_dim: number of outputs (10 returns)
    dropout   : dropout rate
    """

    inputs = Input(shape=(seq_len, input_dim))  # (None, 30, 80)

    # Step A: Project from 80 -> hidden_dim (e.g., 128) for the Transformer
    x = layers.Dense(hidden_dim, activation='relu')(inputs)

    # Step B: Stack multiple Transformer blocks
    for _ in range(num_blocks):
        x = TransformerBlock(embed_dim=hidden_dim,
                             num_heads=num_heads,
                             ff_dim=ff_dim,
                             dropout=dropout)(x)

    # Step C: Take the final time step or do pooling
    x = x[:, -1, :]  # => (batch, hidden_dim)

    # Step D: Final Dense => output_dim (10 returns)
    outputs = layers.Dense(output_dim)(x)

    model = Model(inputs, outputs)
    return model
```

```python
@tf.function
def r2_score(y_true, y_pred):
    y_true_f = tf.reshape(y_true, [-1])
    y_pred_f = tf.reshape(y_pred, [-1])
    ss_res = tf.reduce_sum(tf.square(y_true_f - y_pred_f))
    ss_tot = tf.reduce_sum(tf.square(y_true_f - tf.reduce_mean(y_true_f)))
    return 1.0 - ss_res / (ss_tot + 1e-6)
```

```python
class R2Metric(tf.keras.metrics.Metric):
    def __init__(self, name='r2_score', **kwargs):
        super().__init__(name=name, **kwargs)
        self.r2_accum = self.add_weight(name='r2_accum', initializer='zeros')
        self.count = self.add_weight(name='count', initializer='zeros')

    def update_state(self, y_true, y_pred, sample_weight=None):
        r2_val = r2_score(y_true, y_pred)
        self.r2_accum.assign_add(r2_val)
        self.count.assign_add(1.0)

    def result(self):
        return self.r2_accum / (self.count + 1e-7)

    def reset_states(self):
        self.r2_accum.assign(0.0)
        self.count.assign(0.0)
```

```python
def train_deeper_transformer(Train_X, Train_Y, Test_X, Test_Y, epochs=20, batch_size=32):
    """
    Example usage with shapes:
      Train_X: (2135, 30, 80)
      Train_Y: (2135, 10)
      Test_X : (330, 30, 80)
      Test_Y : (330, 10)
    """
    model = build_deeper_transformer(
        seq_len=Train_X.shape[1],   # 30
        input_dim=Train_X.shape[2], # 80
        hidden_dim=128,             # bigger dimension
        num_heads=4,
        ff_dim=256,
        num_blocks=3,              # deeper: 3 blocks
        output_dim=Train_Y.shape[1],
        dropout=0.1
    )

    r2_metric = R2Metric()

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
        loss='mse',
        metrics=['mae', r2_metric]
    )

    model.summary()

    # Train
    model.fit(
        Train_X, Train_Y,
        validation_data=(Test_X, Test_Y),
        epochs=epochs,
        batch_size=batch_size
    )

    # Evaluate
    loss, mae, r2 = model.evaluate(Test_X, Test_Y)
    print(f"[Test] Loss={loss:.4f}, MAE={mae:.4f}, R²={r2:.4f}")

    preds = model.predict(Test_X)
    return model,preds
```
